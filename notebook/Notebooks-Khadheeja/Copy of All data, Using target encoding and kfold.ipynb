```python
# Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
```

Feature Engineering

```python
# STRATIFIES K-FOLD CROSS VALIDATION { 10-fold }

# Import Required Modules.
from statistics import mean, stdev
from sklearn import preprocessing
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn import datasets
from sklearn.metrics import r2_score

y_data=data['total_price_log']
X_data=data.drop(['total_price_log','total_price'],axis=1)
# Feature Scaling for input features.
#scaler = preprocessing.MinMaxScaler()
#X_scaled = scaler.fit_transform(X)


# Create StratifiedKFold object.
kf = KFold(n_splits=10, shuffle=True, random_state=36)
lst_r2_stratified = []

for train_index, test_index in kf.split(X_data, y_data):
    # Use `.iloc` to properly select rows by position
    X_train_fold = X_data.iloc[train_index].copy()
    X_test_fold = X_data.iloc[test_index].copy()
    y_train_fold = y_data.iloc[train_index].copy()
    y_test_fold = y_data.iloc[test_index].copy()
    # Handle target
    shape_means = X_train_fold.groupby('shape')['price_per_carat_log'].mean()
    origin_means = X_train_fold.groupby('origin')['price_per_carat_log'].mean()
    color_means = X_train_fold.groupby('color')['price_per_carat_log'].mean()
    color_intensity_means = X_train_fold.groupby('color_intensity')['price_per_carat_log'].mean()
    clarity_means = X_train_fold.groupby('clarity')['price_per_carat_log'].mean()
    treatment_means = X_train_fold.groupby('treatment')['price_per_carat_log'].mean()
    cut_means = X_train_fold.groupby('cut')['price_per_carat_log'].mean()

    # Map target encoding to the training data
    X_train_fold['shape_encoded'] = X_train_fold['shape'].map(shape_means)
    X_train_fold['origin_encoded'] = X_train_fold['origin'].map(origin_means)
    X_train_fold['color_encoded'] = X_train_fold['color'].map(color_means)
    X_train_fold['color_intensity_encoded'] = X_train_fold['color_intensity'].map(color_intensity_means)
    X_train_fold['clarity_encoded'] = X_train_fold['clarity'].map(clarity_means)
    X_train_fold['cut_encoded'] = X_train_fold['cut'].map(cut_means)
    X_train_fold['treatment_encoded'] = X_train_fold['treatment'].map(treatment_means)

    # Map target encoding to the test data
    # Handle unseen categories by filling with global mean
    global_mean = X_train_fold['price_per_carat_log'].mean()
    X_test_fold['shape_encoded'] = X_test_fold['shape'].map(shape_means).fillna(global_mean)
    X_test_fold['origin_encoded'] = X_test_fold['origin'].map(origin_means).fillna(global_mean)
    X_test_fold['color_encoded'] = X_test_fold['color'].map(color_means).fillna(global_mean)
    X_test_fold['color_intensity_encoded'] = X_test_fold['color_intensity'].map(color_intensity_means).fillna(global_mean)
    X_test_fold['clarity_encoded'] = X_test_fold['clarity'].map(clarity_means).fillna(global_mean)
    X_test_fold['cut_encoded'] = X_test_fold['cut'].map(cut_means).fillna(global_mean)
    X_test_fold['treatment_encoded'] = X_test_fold['treatment'].map(treatment_means).fillna(global_mean)


    print(X_train_fold.dtypes)
    X_train_fold = X_train_fold.drop(columns=['price_per_carat_log', 'color', 'shape', 'color_intensity', 'origin', 'cut', 'treatment', 'clarity','type','carat'])
    X_test_fold = X_test_fold.drop(columns=['price_per_carat_log', 'color', 'shape', 'color_intensity', 'origin', 'cut', 'treatment', 'clarity','type','carat'])

    # Train the model.
    rfr = RandomForestRegressor(random_state=36, n_estimators=25)
    rfr.fit(X_train_fold, y_train_fold)

    # Predictions
    y_pred_fold=rfr.predict(X_test_fold)
    print(y_pred_fold)

    y_pred_fold_original = np.expm1(y_pred_fold)
    y_test_fold_original = np.expm1(y_test_fold)

    # Transform predictions back to original state
    print("Predictions:", y_pred_fold_original)
    print("Ground Truth: ", y_test_fold_original)

    # Handle invalid values
    y_pred_fold_original = np.nan_to_num(y_pred_fold_original, nan=0.0, posinf=1e10, neginf=-1e10)
    y_test_fold_original = np.nan_to_num(y_test_fold_original, nan=0.0, posinf=1e10, neginf=-1e10)


    # Evaluate the model
    r2 = r2_score(y_test_fold_original,y_pred_fold_original)
    lst_r2_stratified.append(r2)

# Print the output.
print('List of possible accuracy:', lst_r2_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:',
	max(lst_r2_stratified)*100, '%')
print('\nMinimum Accuracy:',
	min(lst_r2_stratified)*100, '%')
print('\nOverall Accuracy:',
	mean(lst_r2_stratified)*100, '%')
print('\nStandard Deviation is:', stdev(lst_r2_stratified))
```
